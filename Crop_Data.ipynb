```python
import glob
import shutil
import os
from PIL import Image

Car_Models = ['Alto', 'Baleno', 'Dzire', 'Eeco', 'Ertiga', 'Swift', 'Vitara Brezza']

for n in range(2*len(Car_Models)):
    curr = n
    if n<7:
        train_valid_path = 'Small_Data_New_train'
    else:
        train_valid_path = 'Small_Data_New_valid'
        curr = curr-7
    print('exp'+('' if n==0 else str(n+1)))
    print(len(glob.glob('./runs/detect/exp'+('' if n==0 else str(n+1))+'/labels/*.txt')))

    for i in glob.glob('./runs/detect/exp'+('' if n==0 else str(n+1))+'/labels/*.txt'):
        if not os.path.isfile(i):
            continue
        txt_file = open(i, 'r')
        detections = txt_file.readlines()
        for count, k in enumerate(detections):
            object_class, norm_x_mid, norm_y_mid, norm_w, norm_h, conf = k.split(' ')
            norm_x1 = float(norm_x_mid) - (float(norm_w)/2)
            norm_y1 = float(norm_y_mid) - (float(norm_h)/2)
            norm_x2 = float(norm_x_mid) + (float(norm_w)/2)
            norm_y2 = float(norm_y_mid) + (float(norm_h)/2)
            norm_w = float(norm_w)
            norm_h = float(norm_h)
            img_path = glob.glob('../../input/car-models-fewshot-to-be-grayscaled/'+train_valid_path+'/'+'Maruti Suzuki '+Car_Models[curr]+'/'+i.split('/')[-1][:-4]+'.*')[0]
            img = Image.open(img_path)
            w, h = img.size
            print('../'+train_valid_path+'/'+'Maruti Suzuki '+Car_Models[curr]+'/'+str(count)+'_'+i.split('/')[-1][:-4]+'.png') 
            img.crop((int(norm_x1*w), int(norm_y1*h), int(norm_x2*w), int(norm_y2*h))).convert('RGB').save('../'+train_valid_path+'/'+'Maruti Suzuki '+Car_Models[curr]+'/'+str(count)+'_'+i.split('/')[-1][:-4]+'.png')
#             print('../'+train_valid_path+'/'+'Maruti Suzuki '+Car_Models[curr]+'/'+str(count)+'_'+i.split('/')[-1][:-4]+'.png')
        txt_file.close()
```
